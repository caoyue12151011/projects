## The problem
In what phase does Venus reach its brightest observed from the Earth? Assuming
that all the planetary orbits are circles.

```python
import numpy as np 
import sympy as sp
import matplotlib.pyplot as plt
import ipywidgets as widgets

# change default matplotlib fonts
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams['figure.dpi'] = 100
```

#### Derived variables (as functions of $a$, $a_0$, $\theta$, $S$)
$a$: orbital semi-major of the observer's planet (OP)\
$a_0$: orbital semi-major of the target planet (TP)\
$\theta$: the OP-Sun-TP angle\
$S$: area of the big circle of the TP globe

```python
f_d = lambda a,a0,theta: (a**2+a0**2-2*a*a0*np.cos(theta))**.5 # OP-TP distance

# OP-TP-Sun angle, the "phase" angle
# 0=full, pi/2=left half (bright), pi=crescent, -pi/2=right half
f_alpha=lambda a,a0,d:np.arccos((a**2+d**2-a0**2)/(2*a*d))*np.sign(np.pi-theta)
    
# solid angle/S of TP's bright part , proportional to TP's brightness    
f_Omega = lambda a,a0,d: ((a+d)**2-a0**2)/(4*a*d**3) 
    
# theta where the TP reaches its brightest, r_a = a/a0
def f_theta_max(r_a):
    Res = np.array([(r_a<=.25)*np.pi,(1<=r_a)*0.,
        (.25<r_a)*(r_a<1)*np.arccos(2*(r_a**2+3)**.5-2*r_a-1/r_a)])
    return np.nansum(Res,axis=0)
```

## Demonstration

```python
a0 = 1
a = .7
theta = np.linspace(0,2*np.pi,500)

r_a = a/a0
theta_d = theta*180/np.pi
d = f_d(a,a0,theta)
alpha = f_alpha(a,a0,d)
Omega = f_Omega(a,a0,d)
theta_max = f_theta_max(r_a)


plt.figure(figsize=(5,4))
plt.plot(theta_d,d,label='d')
plt.plot(theta_d,alpha,label='alpha')
plt.plot(theta_d,Omega,label='Omega')
plt.legend()
plt.grid()
plt.xlabel(r'$\theta\ (^\circ)$')
plt.ylabel(r'$\Omega_{\rm bright}$')
plt.tight_layout()
plt.show()
```

```python
# variables
a0 = 1 
A = np.linspace(.05,2,400) # array of a
Theta = np.linspace(0,2*np.pi,500) # array of theta

# derived variables
R_a = A/a0
Theta_d = Theta*180/np.pi



def demo_plot(a0,a,theta_d):
    fig, axes = plt.subplots(2,2,figsize=(5.5,4.5))
    ax1 = axes[0,0]
    ax2 = axes[0,1]
    ax3 = axes[1,0]
    ax4 = axes[1,1]
    
    # derivation
    theta = theta_d*np.pi/180
    r_a = a/a0

    D = f_d(a,a0,Theta)
    Br = f_Omega(a,a0,D)
    fct = np.nanmax(Br)
    Br /= fct # normalization
    Theta_max = f_theta_max(R_a)
    Theta_max_d = Theta_max*180/np.pi
        
    d = f_d(a,a0,theta)
    br = f_Omega(a,a0,d)
    br /= fct
    theta_max = f_theta_max(r_a)
    theta_max_d = theta_max*180/np.pi


    # theta vs brightness plot
    ax1.plot(Theta_d,Br,color='k',linewidth=1)
    ax1.scatter(theta_d,br,s=5,facecolor='r',edgecolor='none')
    ax1.scatter(theta_d,br,s=50,facecolor='r',edgecolor='none',alpha=.3)
    ax1.grid()
    ax1.xaxis.set_label_text(r'$\theta\ (^\circ)$')
    ax1.yaxis.set_label_text('Rel. brightness')


    # r_a vs theta_max plot
#     plt.subplot(2,2,3)
#     plt.plot(R_a,Theta_max_d,color='k',linewidth=1)
#     plt.scatter(r_a,theta_max_d,s=5,facecolor='r',edgecolor='none')
#     plt.scatter(r_a,theta_max_d,s=50,facecolor='r',edgecolor='none',alpha=.3)
#     plt.grid()
#     plt.xlabel(r'$a/a_0$')
#     plt.ylabel(r'$\theta_{\rm max}\ (^\circ)$')
    
    
    # orbit plot
#     plt.subplot(2,2,2)
    plt.show()

w=widgets.interact(demo_plot,
                 a0=widgets.fixed(a0),a=(0,2.5,.01),theta_d=(0,360,.5))
# plt.tight_layout()

```
